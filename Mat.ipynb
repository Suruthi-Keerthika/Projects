                                                         LINE PLOT

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[4,8,12,16,20]
plt.plot(x,y)
plt.title("Cost of Eggs")
plt.xlabel("No.of.Eggs")
plt.ylabel("Costs in Rs")
plt.show()
```

```python
z=[6,12,18,24,30]
plt.plot(x,y)
plt.plot(x,z)
plt.title("Cost of Eggs")
plt.xlabel("No.of.Eggs")
plt.ylabel("Costs in Rs")
plt.show()
```

```python
plt.plot(x,y, label='Hen', color='r', linestyle='--', marker='o')
plt.plot(x,z, label='Quail', color='b', linestyle='-', marker='.')
plt.title("Cost of Eggs")
plt.xlabel("No.of.Eggs")
plt.ylabel("Costs in Rs")
plt.legend()
plt.show()
```

```python
plt.style.use('ggplot')
plt.plot(x,y, label='Hen', color='r', linestyle='--', marker='o')
plt.plot(x,z, label='Quail', color='b', linestyle='-', marker='.')
plt.title("Cost of Eggs")
plt.xlabel("No.of.Eggs")
plt.ylabel("Costs in Rs")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

                                                               BAR CHART

```python
x=['A', 'B', 'C', 'D']
y=[100,200,150,250]
plt.bar(x,y)
plt.title('Height of the Buildings')
plt.show()
```

                                                          PIE-CHART

```python
labels=['School', 'Home', 'Play', 'Tuition', 'Others']
sizes=[8, 10, 2, 3, 1]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Daily Routine hours")
plt.show()
```
