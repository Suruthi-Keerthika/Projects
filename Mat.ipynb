                                                         LINE PLOT

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[4,8,12,16,20]
plt.plot(x,y)
plt.title("Cost of Eggs")
plt.xlabel("Eggs")
plt.ylabel("Costs in Rs")
plt.show()
```

                                                               BAR CHART

```python
x=['A', 'B', 'C', 'D']
y=[100,200,150,250]
plt.bar(x,y)
plt.title('Height of the Buildings')
plt.show()
```

                                                          PIE-CHART

```python
labels=['School', 'Home', 'Play', 'Tuition', 'Others']
sizes=[8, 10, 2, 3, 1]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Daily Routine hours")
plt.show()
```

                                                                HISTOGRAM

```python
data=[13,22,17,34,23,31,18,27,10,26,40,56,84,48,64,94]
plt.hist(data, bins=15)
plt.title("HISTOGRAM")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
