```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[4,8,12,16,20]
plt.plot(x,y)
plt.title("Cost of Eggs")
plt.xlabel("No.of.Eggs")
plt.ylabel("Costs in Rs")
plt.show()
```

```python
z=[6,12,18,24,30]
plt.plot(x,y)
plt.plot(x,z)
plt.title("Cost of Eggs")
plt.xlabel("No.of.Eggs")
plt.ylabel("Costs in Rs")
plt.show()
```

```python
plt.plot(x,y, label='Hen', color='r', linestyle='--', marker='o')
plt.plot(x,z, label='Quail', color='b', linestyle='-', marker='.')
plt.title("Cost of Eggs")
plt.xlabel("No.of.Eggs")
plt.ylabel("Costs in Rs")
plt.legend()
plt.show()
```

```python
plt.style.use('ggplot')
plt.plot(x,y, label='Hen', color='r', linestyle='--', marker='o')
plt.plot(x,z, label='Quail', color='b', linestyle='-', marker='.')
plt.title("Cost of Eggs")
plt.xlabel("No.of.Eggs")
plt.ylabel("Costs in Rs")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

                                                               BAR CHART

```python
x=['A', 'B', 'C', 'D']
y=[100,200,150,250]
plt.bar(x,y, color='r', label="Others")
plt.title('Height of the Buildings')
plt.show()
```

```python
x=['A', 'B', 'C', 'D']
ap_y=[300,279,350,420]
plt.bar(x,ap_y, color='y',label="Apartment")
hl_y=[150,128,320,256]
plt.bar(x,hl_y, color='b', label="Hospital")
y=[100,200,150,250]
plt.bar(x,y, color='r', label="Others")
plt.xlabel("Name of the buildings")
plt.ylabel("Height in feets")
plt.legend()
plt.title("Height of the Buildings")
plt.show()
```

```python
import numpy as np
x=['A', 'B', 'C', 'D']
x_indexes=np.arange(len(x))
width=0.25
ap_y=[300,279,350,420]
plt.bar(x_indexes-width,ap_y, width=width,  color='y', label="Apartments")
hl_y=[150,128,320,256]
plt.bar(x_indexes,hl_y, width=width,  color='b',label="Hospitals")
y=[100,200,150,250]
plt.bar(x_indexes+width, y, width=width,  color='r',label="Others")
plt.xlabel("Name of the buildings")
plt.ylabel("Height in feets")
plt.legend()
plt.title("Height of the Buildings")
plt.tight_layout()
plt.show()
```

                                                          PIE-CHART

```python
labels=['School', 'Home', 'Play', 'Tuition', 'Others']
sizes=[8, 10, 2, 3, 1]
plt.pie(sizes, labels=labels)
plt.title("Daily Routine hours")
plt.show()
```

```python
labels=['School', 'Home', 'Play', 'Tuition', 'Others']
colors=['Red','Yellow','Green','Brown','Blue']
sizes=[8, 10, 2, 3, 1]
plt.pie(sizes, labels=labels, colors=colors, wedgeprops={'edgecolor':'Black'})
plt.title("Daily Routine hours")
plt.show()
```

```python
labels=['School', 'Home', 'Play', 'Tuition', 'Others']
colors=['Red','Yellow','Green','Brown','Blue']
explode=[0,0,0.2,0,0]
sizes=[8, 10, 2, 3, 1]
plt.pie(sizes, labels=labels, colors=colors,explode=explode, wedgeprops={'edgecolor':'Black'},autopct='%1.1f%%')
plt.title("Daily Routine hours")
plt.show()
```

```python
labels=['School', 'Home', 'Play', 'Tuition', 'Others']
colors=['Red','Yellow','Green','Brown','Blue']
sizes=[8, 10, 2, 3, 1]
plt.pie(sizes, labels=labels, colors=colors, wedgeprops={'edgecolor':'Black'}, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Daily Routine hours")
plt.show()
```
